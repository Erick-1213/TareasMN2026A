## Pregunta 1

Use el método de bisección para encontrar soluciones precisas dentro de $10^{-2}$ para $x^3 - 7x^2 + 14x - 6 = 0$ en cada intervalo.

**a.** $[0, 1]$  
**b.** $[1, 3.2]$  
**c.** $[3.2, 4]$

```python
from scipy.optimize import bisect

def f(x):
    return x**3 - 7*x**2 + 14*x - 6

intervalos = [(0, 1), (1, 3.2), (3.2, 4)]
tolerancia = 1e-2

for a, b in intervalos:
    raiz = bisect(f, a, b, xtol=tolerancia)
    print(f"Raíz en el intervalo [{a}, {b}]: {raiz:.5f}")
```

## Pregunta 2

a. Dibuje las gráficas para $𝑦 = 𝑥 y 𝑦 = sin 𝑥$. 

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

y1 = x
y2 = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label="y = x", color="blue")
plt.plot(x, y2, label="y = sin(x)", color="red")

plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1)

plt.title("Gráficas de y = x y y = sin(x)")
plt.xlabel("x (radianes)")
plt.ylabel("y")

plt.legend()
plt.grid(True)

plt.show()
```

**b.** Use el método de bisección para encontrar soluciones precisas dentro de $10^{-5}$ para el primer valor positivo de $x$ con $x = 2 \sin x$.

```python
def g(x):
    return x - 2 * np.sin(x)

intervalo = (1, 5)
tolerancia = 1e-5

raiz = bisect(g, *intervalo, xtol=tolerancia)

print(f"Raíz aproximada en el intervalo {intervalo}: {raiz:.6f}")
```

## Pregunta 3

a. Dibuje las gráficas para $𝑦 = 𝑥 y 𝑦 = \tan 𝑥. $

```python
import numpy as np
import matplotlib.pyplot as plt

valores = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

linea_recta = valores
funcion_tan = np.tan(valores)

funcion_tan[np.abs(funcion_tan) > 10] = np.nan

plt.figure(figsize=(8, 6))
plt.plot(valores, linea_recta, label='y = x', color="#A87FC6")
plt.plot(valores, funcion_tan, label='y = tan(x)', color="#3A33FF")

plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Comparación entre y = x y y = tan(x)')
plt.legend()
plt.grid(True)
plt.show()
```
