### Pregunta 4

Sea $f(x) = (x+3)(x+1)^2x(x-1)^3(x-3).$ ¿ En qué cero de $f$ converge el método de bisección cuando se aplica en los siguientes intervalos?

a. $[-1.5,2.5]$

b. $[-0.5,2.4]$

c. $[-0.5,3]$

d. $[-3,-0.5]$

```python
def biseccion(f, a, b, TOL, N=1000):
    FA = f(a)
    FB = f(b)

    if abs(FA) < TOL:
        return a
    if abs(FB) < TOL:
        return b
    if FA * FB > 0:
        print(f"No hay cambio de signo en el intervalo [{a}, {b}]")
        return None

    for _ in range(N):
        p = (a + b) / 2
        FP = f(p)
        if abs(b - a) / 2 < TOL or abs(FP) < TOL:
            return p
        if FA * FP > 0:
            a, FA = p, FP
        else:
            b = p
    print(f"No se alcanzó la tolerancia en {N} iteraciones.")
    return None


def f6(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)


intervalos = [
    (-1.5, 2.5),
    (-0.5, 2.4),
    (-0.5, 3),
    (-3, -0.5)
]

for idx, (a, b) in enumerate(intervalos, start=1):
    raiz = biseccion(f6, a, b, 1e-3)
    print(f"Intervalo {idx}) [{a}, {b}] → Raíz aproximada: {raiz}")
```
